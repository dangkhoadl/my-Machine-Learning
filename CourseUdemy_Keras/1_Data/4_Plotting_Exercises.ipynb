## Exercise 2
- load the dataset: `../data/weight-height.csv`
- inspect it
- plot it using a scatter plot with Weight as a function of Height
- plot the male and female populations with 2 different colors on a new scatter plot
- remember to label the axes

```python
# load the dataset: ../data/weight-height.csv
df = pd.read_csv('weight-height.csv')

display(df.info())
display(df.head())
display(df.describe())
display(df['Gender'].value_counts())

# plot it using a scatter plot with Weight as a function of Height
df.plot(
    kind='scatter',
    x='Height', y='Weight')
```

```python
# plot the male and female populations with 2 different colors on a new scatter plot - 1
males = df[df['Gender'] == 'Male']
females = df.query('Gender == "Female"')

fig, ax = plt.subplots()

males.plot(kind='scatter', x='Height', y='Weight',
           ax=ax, color='blue', alpha=0.3,
           title='Male & Female Populations')

females.plot(kind='scatter', x='Height', y='Weight',
             ax=ax, color='red', alpha=0.3)
```

```python
# plot the male and female populations with 2 different colors on a new scatter plot - 2
df['Gendercolor'] = df['Gender'].map({'Male': 'blue', 'Female': 'red'})
display(df.head())

df.plot(
    kind='scatter', 
    x='Height', y='Weight',
    c=df['Gendercolor'],
    alpha=0.3,
    title='Male & Female Populations')
```

```python
# plot the male and female populations with 2 different colors on a new scatter plot - 3
fig, ax = plt.subplots()

ax.plot(
    males['Height'], males['Weight'], 'ob', 
    females['Height'], females['Weight'], 'or',
    alpha=0.3)

plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Male & Female Populations')
```

## Exercise 3
- plot the histogram of the heights for males and for females on the same plot
- use alpha to control transparency in the plot comand
- plot a vertical line at the mean of each population using `plt.axvline()`

```python
males['Height'].plot(kind='hist',
                     bins=50,
                     range=(50, 80),
                     alpha=0.3,
                     color='blue')

females['Height'].plot(kind='hist',
                       bins=50,
                       range=(50, 80),
                       alpha=0.3,
                       color='red')

plt.title('Height distribution')
plt.legend(["Males", "Females"])
plt.xlabel("Heigth (in)")


plt.axvline(males['Height'].mean(), color='blue', linewidth=2)
plt.axvline(females['Height'].mean(), color='red', linewidth=2)
```

```python
males['Height'].plot(kind='hist',
                     bins=200,
                     range=(50, 80),
                     alpha=0.3,
                     color='blue',
                     cumulative=True,
                     normed=True)

females['Height'].plot(kind='hist',
                       bins=200,
                       range=(50, 80),
                       alpha=0.3,
                       color='red',
                       cumulative=True,
                       normed=True)

plt.title('Height distribution')
plt.legend(["Males", "Females"])
plt.xlabel("Heigth (in)")

plt.axhline(0.8)
plt.axhline(0.5)
plt.axhline(0.2)
```
