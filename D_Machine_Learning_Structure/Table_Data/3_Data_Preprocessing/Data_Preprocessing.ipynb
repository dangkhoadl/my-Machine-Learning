```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Read data

```python
filename = "pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pd.read_csv(filename, names=names)

data.head(20)
```

```python
data.shape
```

```python
data.dtypes
```

### Separate dataset

```python
data = data.values
```

```python
X = data[:, 0:8]
Y = data[:, 8]
```

### Rescale Data
- normalization = rescaled into the range between 0 and 1
- useful for optimization algorithms used in the core of machine learning algorithms like gradient descent
- also useful for algorithms that weight inputs like regression and neural networks and algorithms that use distance measures like k -Nearest Neighbors

```python
from sklearn.preprocessing import MinMaxScaler

# Rescale data (between 0 and 1)
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
```

```python
# summarize transformed data
np.set_printoptions(precision=3)

print rescaledX[0:5,:]
```

### Standardize Data
- a useful technique to transform attributes with a Gaussian distribution and differing means and standard deviations to a standard Gaussian distribution with a mean of 0 and a standard deviation of 1
- most suitable for techniques that assume a Gaussian distribution in the input variables and work better with rescaled data
    - linear regression
    - logistic regression
    - linear discriminate analysis

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)
```

```python
# summarize transformed data
np.set_printoptions(precision=3)
print(rescaledX[0:5,:])
```

### Normalize Data
- rescaling each observation (row) to have a length of 1 (called a unit norm or a vector with the length of 1 in linear algebra)
- useful for sparse datasets (lots of zeros) with attributes of varying scales when using algorithms that weight input values
    + neural networks
    + k -Nearest Neighbors

```python
from sklearn.preprocessing import Normalizer

scaler = Normalizer().fit(X)
normalizedX = scaler.transform(X)
```

```python
# summarize transformed data
np.set_printoptions(precision=3)
print(normalizedX[0:5,:])
```

### Binarize Data (Make Binary)
- transform data using a binary threshold
- All values above the threshold are marked 1 and all equal to or below are marked as 0
- useful when feature engineering and you want to add new features that indicate something meaningful

```python
from sklearn.preprocessing import Binarizer

binarizer = Binarizer(threshold=0.0).fit(X)
binaryX = binarizer.transform(X)
```

```python
# summarize transformed data
np.set_printoptions(precision=3)
print(binaryX[0:5,:])
```
