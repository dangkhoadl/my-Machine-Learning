## White noise
- A sequence of random numbers and cannot be predicted
- A time series is white noise if
    + the variables are independent and identically distributed with a mean of zero
    + This means that all variables have the same variance and each value has a zero correlation with all other values
-  If your time series is white noise, then, by definition, it is random -> Cannot predict

## Is your Time Series White Noise?
- Does your series have a non-zero mean?
- Does the variance change over time?
- Do values correlate with lag values?


## How to check if time series is whitenoise ?
- __Create a line plot__: Check for gross features like changing:
	+ mean
	+ variance
	+ relationship between lagged variables
- __Caculate summary Stats__: Check the mean and variance of the whole series ><  mean and variance of meaningful contiguous blocks of values in the series (eg:days, months, years)
- __Create an autocorrelation plot__: Check for gross correlation between lagged variables

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from random import gauss
from random import seed
from pandas.plotting import autocorrelation_plot

# seed random number generator
seed(1)

# create white noise series(Gaussian mean = 0.0, std = 1.0)
series = pd.Series([gauss(0.0, 1.0) for i in range(1000)])

# summary stats
series.describe()
```

### Plot

```python
series.plot();
```

```python
series.plot(kind='density');
```

```python
# autocorrelation
autocorrelation_plot(series)
plt.show()
```
