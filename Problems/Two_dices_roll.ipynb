```python
from numpy.random import randint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

## Calculate by Probability

```python
from itertools import product

# Possibility in range [2,12]
results = dict.fromkeys(range(2,13), 0)

# Generate all possible outcomes
outcomes = list(product(range(1,7), range(1,7)))
sample_size = len(outcomes)
for (d1,d2) in outcomes:
    results[d1+d2] += 1

# Print
print(results)
print("Sample size = {}".format(sample_size))

# Plot
fig, ax = plt.subplots()
data_plot = ax.bar(results.keys(), results.values(), color='g')

autolabel(data_plot)
plt.xticks(np.arange(2,13,1))

plt.show()
```

```python
# Probability
prob = {key: round(value*100.0/sample_size, 2) for (key, value) in results.items()}

# Print
print(prob)

# Plot
fig, ax = plt.subplots()
data_plot = ax.bar(prob.keys(), prob.values(), color='g')

plt.xlabel('X', fontsize=18)
plt.ylabel('P(X)', fontsize=18)
autolabel(data_plot)
plt.xticks(np.arange(2,13,1))

plt.show()
```

## Calculate by Monte Carlo Simulation

```python
sample_size = int(1e6)
```

```python
# Possibility in range [2,12]
results = dict.fromkeys(range(2,13), 0)

for i in range(sample_size):
    d1 = randint(1,7)
    d2 = randint(1,7)
    results[d1+d2] += 1

# Plot
fig, ax = plt.subplots()
data_plot = ax.bar(results.keys(), results.values(), color='g')

autolabel(data_plot)
plt.xticks(np.arange(2,13,1))

plt.show()
```

```python
# Probability
prob = {key: round(value*100.0/sample_size, 2) for (key, value) in results.items()}

# Print
print(prob)

# Plot
fig, ax = plt.subplots()

data_plot = ax.bar(prob.keys(), prob.values(), color='g')

autolabel(data_plot)
plt.xlabel('X', fontsize=18)
plt.ylabel('P(X)', fontsize=18)
plt.xticks(np.arange(2,13,1))

plt.show()
```
