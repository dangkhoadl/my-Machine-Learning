# Read csv file

```python
# df = pd.read_csv('input.csv')
df = pd.read_csv('olympics.csv', index_col = 0, skiprows=1)
df.head()
```

```python
# Rename
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold' + col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#' + col[1:]}, inplace=True) 
df.head()
```

# Querying a DataFrame

```python
# # Query countries that have gold in Summer
only_gold = df.where(df['Gold'] > 0)
only_gold.head()
```

```python
# Drop nan rows
only_gold = only_gold.dropna()
only_gold.head()
```

# Query DataFrame

```python
df = pd.read_csv('census.csv')
df.head()
```

```python
# Query Sumlev = 50
df=df[df['SUMLEV'] == 50]
df.head()
```

```python
# Refilt columns
columns_to_keep = ['STNAME',
                   'CTYNAME',
                   'BIRTHS2010',
                   'BIRTHS2011',
                   'BIRTHS2012',
                   'BIRTHS2013',
                   'BIRTHS2014',
                   'BIRTHS2015',
                   'POPESTIMATE2010',
                   'POPESTIMATE2011',
                   'POPESTIMATE2012',
                   'POPESTIMATE2013',
                   'POPESTIMATE2014',
                   'POPESTIMATE2015']
df = df[columns_to_keep]
df.head()
```

```python
# Reindex
df = df.set_index(['STNAME', 'CTYNAME'])
df.head()
```
